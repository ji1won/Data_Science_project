### 정규분포가 아닌 원분포를 보고 수업시간에 배운 Central Limit Theorem (중심극한정리)를 확인해보고자 함.

```python
count = time_count_table.select('Count')

sample_size = 400
repetitions = 10000

means = make_array()

for i in np.arange(repetitions):
    sample = count.sample(sample_size)
    new_mean = np.mean(sample.column('Count'))
    means = np.append(means, new_mean)

results = Table().with_column(
    'Sample Mean', means
)
```

```python
results.hist(bins=np.arange(1500, 2000, 30))
```

 원 분포와 상관없이 샘플의 평균은 정규분포가 됨을 확인할 수 있었다.

#이번에는 시간과 발생 빈도의 상관관계를 파악해보자

```python
time_count_table.scatter('시간', 'Count')
```

양을 경향성을 보이는 듯하나 20시간이 넘어가면서 오히려 떨어짐

데이터 사이의 차이를 일반화시킬 수 있도록 standard unit으로 바꿔서 진행해 봄

```python
def standard_units(any_numbers):
    "Convert any array of numbers to standard units."
    return (any_numbers - np.mean(any_numbers))/np.std(any_numbers)  #Z변환
```

```python
Table().with_columns(
    '시간 (standard units)',  standard_units(time_count_table.column('시간')),
    'Count (standard units)', standard_units(time_count_table.column('Count'))
).scatter(0, 1)
plots.xlim(-3, 3)
plots.ylim(-3, 3);
```

```python
def correlation(t, label_x, label_y):
    return np.mean(standard_units(t.column(label_x))*standard_units(t.column(label_y)))

def slope(t, label_x, label_y):
    r = correlation(t, label_x, label_y)
    return r*np.std(t.column(label_y))/np.std(t.column(label_x))

def intercept(t, label_x, label_y):
    return np.mean(t.column(label_y)) - slope(t, label_x, label_y)*np.mean(t.column(label_x))
```

```python
time_count_r = correlation(time_count_table, '시간', 'Count')
time_count_r
```

Correlation은 Linear Association을 의미하는것을 고려하면 어느정도의 선형적 관계를 가진다고 판단할 수 있다.

```python
time_count_slope = slope(time_count_table, '시간', 'Count')
time_count_intercept = intercept(time_count_table, '시간', 'Count')
time_count_slope, time_count_intercept
```

```python
count_with_predictions = time_count_table.with_column(
    'Regression Prediction', time_count_slope*time_count_table.column('시간') + time_count_intercept
)
count_with_predictions
```

```python
count_with_predictions.scatter('시간')
```

regression line을 통해 prediction을 수행할 수 가능
시간이 1시간 증가시 교통사고 빈도는 0.671 횟수만큼 증가한다

#regression diagnostics

```python
def fit(table, x, y):
    """Return the height of the regression line at each x value."""
    a = slope(table, x, y)
    b = intercept(table, x, y)
    return a * table.column(x) + b
```

```python
def residual(table, x, y):
    return table.column(y) - fit(table, x, y)
```

```python
time_count_table = time_count_table.with_columns(
        'Fitted Value', fit(time_count_table, '시간', 'Count'),
        'Residual', residual(time_count_table, '시간', 'Count')
    )
time_count_table
```

```python
def scatter_fit(table, x, y):
    table.scatter(x, y, s=15)
    plots.plot(table.column(x), fit(table, x, y), lw=4, color='gold')
    plots.xlabel(x)
    plots.ylabel(y)
```

```python
scatter_fit(time_count_table, '시간', 'Count')
```

+ $x$와 $residual$ 사이의 분포 상관성을 그려보자.

```python
def residual_plot(table, x, y):
    x_array = table.column(x)
    t = Table().with_columns(
            x, x_array,
            'residuals', residual(table, x, y)
        )
    t.scatter(x, 'residuals', color='r')
    xlims = make_array(min(x_array), max(x_array))
    plots.plot(xlims, make_array(0, 0), color='darkblue', lw=4)
    plots.title('Residual Plot')
```

```python
residual_plot(time_count_table, '시간', 'Count')
```
